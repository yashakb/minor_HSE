# Метрики качества классификации

## Точность и полнота

Пусть мы решаем задачу классификации, т.е. нам задана обучающая выборка $\left(x_i, y_i \right)_{i=1}^\ell$, где $y_i$ — значения из некоторого конечного множества.

Ранее при решении задач классификации за метрику качества мы принимали accuracy:
$$accuracy = \sum_{i=1}^\ell [a(x_i) = y_i]$$

Однако accuracy как метрика качества в задачах классификации имеет ряд проблем, в частности:
* имеет неадекватные значения на задачах с несбалансированными классами;
* не учитывает цены ошибок.

Рассмотрим две другие метрики качества в задачах классификации: **точность** (*precision*) и **полноту** (*recall*). Будем считать, что мы решаем задачу бинарной классификации с метками классов $\mathbb{Y} = \{ -1, +1\}.$

Пусть мы построили классификатор $a(x)$, с помощью которого классифицируем некоторый объект $x_i$ выборки, для которого известно истинное значение $y_i$. При $a(x_i) = +1$ говорят, что классификатор **выделяет объект** $x_i$. Очевидно, возможны четыре случая:
* $a(x_i) = +1, y_i = +1$;
* $a(x_i) = +1, y_i = -1$;
* $a(x_i) = -1, y_i = +1$;
* $a(x_i) = -1, y_i = -1$.

Применим классификатор ко всей выборке и посчитаем, сколько объектов отвечает каждому из этих четырёх исходов. Результат занесём в таблицу, которая часто называется **матрицей ошибок** (*confusion matrix*).
Каждая из четырёх ячеек таблицы имеет своё название. Каждое из названий состоит из двух слов:
* Классификатор ответил верно? (True или False)
* К какому классу классификатор отнёс объект? (Positive или Negative)


| | y = +1 | y = -1 |
|------|------|
|   **a(x) = +1**  | TP| FP |
| **a(x) = -1**| FN | TN |

$precision = \dfrac{TP}{TP + FP}$ — доля в действительности положительных объектов среди выделенных алгоритмом; характеризует то, можем ли мы доверять алгоритму при $a(x) = +1$.

$recall = \dfrac{TP}{TP + FN}$ — доля выделенных алгоритмом объектов среди в действительности положительных; характеризует то, сколько положительных объектов находит классификатор.

Точность без полноты и полнота без точности сами по себе являются довольно бесполезными метриками. Например, пусть «алгоритм» — это врач, класс «+1» — пациент болен, а «−1» — здоров. 
* У ленивого врача, который будет говорить пациенту, что он болен, только если это совсем очевидно, будет иметь место FP = 0, поэтому precision = 1.
* У неуверенного в себе врача, который будет каждому пациенту говорить, что он болен, а потом отправлять на дальнейшее лечение, будет иметь место FN = 0, поэтому recall = 1.

Но профессионализм и того, и другого, сомнителен.

**Задача 1.** Как выглядит матрица ошибок для алгоритма с precision = 1 и recall = 1?

В идеале мы хотим достичь precision = 1 и recall = 1 одновременно, поэтому нужно максимизировать оба значения одновременно.  Для этого используется метрика качества, называемая F-мерой:
$$F = \frac{2 \cdot precision \cdot recall}{precision +recall}$$

F-мера, как и точность и полнота, принимает значения из отрезка [0; 1].

```python
prec = np.linspace (0, 1, 100)
rec = np.linspace(0, 1, 100)
prec_grid, rec_grid = np.meshgrid(prec, rec)
F_grid = 2 * prec_grid * rec_grid / (prec_grid + rec_grid)

fig, ax = plt.subplots(1,1, sharey=True, figsize = (10, 10))
CS = ax.contour(prec_grid, rec_grid, F_grid, 20)
ax.set_xlabel('precision')
ax.set_ylabel('recall')
```

Также существуют вариации, которые специальным образом расставляют приоритеты в паре «точность-полнота».

## Оценки принадлежности классу

Как мы ранее убедились, зачастую в качестве ответа мы получаем не метку класса, а некоторую оценку принадлежности к классу (например, логистическая регрессия выдает вероятность). Тогда итоговый классификатор, выдающий метку класса, имеет следующий вид: $$a(x) = [b(x) > t]$$ где $b(x)$ — ответ классификатора (оценка принадлежности классу +1), $t$ — некоторый фиксированный порог.

Таким образом, мы **бинаризуем** ответ классификатора по порогу $t$. Варьируя значение этого порога, мы можем менять качество итогового классификатора. Возникает несколько вопросов:
* Как выбрать порог $t$?
* Допустим, порог выбран, однако качество работы классификатора $a(x)$ находится на низком уровне. Как понять, проблема состоит в неверном выборе модели $b(x)$ или неверном выборе порога $t$?
* Как оценить качество модели $b(x)$ для предсказания оценок принадлежности, а не итогового классификатора $a(x)$?

Рассмотрим на примере: будем решать задачу кредитного скоринга. По некоторой банковской информации необходимо понять, будет ли испытывать клиент финансовые затруднения в будущем. Данные взяты из [этого](https://www.kaggle.com/c/GiveMeSomeCredit) соревнования. Целевая переменная *SeriousDlqin2yrs* равна 1, если финансовые затруднения будут иметь место, 0 — в противном случае.

```python
data = pd.read_csv('data.csv')
data.head()
```

Можно заметить что выборка весьма несбалансирована:

```python
data['SeriousDlqin2yrs'].value_counts()
```

Попробуем обучить логистическую регрессию и посмотреть, какого качества удаётся достичь. В этом случае, очевидно, $t = 0.5$:

```python
from sklearn.linear_model import LogisticRegression

from sklearn.cross_validation import KFold, cross_val_score, train_test_split

X, y = data.drop(['SeriousDlqin2yrs'], axis=1), data['SeriousDlqin2yrs']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
clf = LogisticRegression(random_state=42)
clf.fit(X_train, y_train)
print('precision:', precision_score(y_test, clf.predict(X_test)))
print('recall:', recall_score(y_test, clf.predict(X_test)))
```

Вероятность, предсказываемая логистической регрессией, принимает значения из отрезка [0; 1]. Давайте будем изменять порог $t$ от 0 до 1, для каждого значения вычислять пару (recall, precision) получившегося алгоритма $a(x)$ и строить для него точку с координатой recall по горизонтали и precision по вертикали. После этого соединим все получившиеся точки — получившаяся кривая называется **PR-кривой** (*precision-recall curve*).

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, clf.predict_proba(X_test)[:, 1])
plt.plot(recall, precision)
plt.xlabel('recall')
plt.ylabel('precision')
plt.show()
```

* PR-кривая проходит через точку (0,0).

* Если при каком-то значении порога $t$ алгоритм $a(x)$ идеально разделяет объекты 2 классов, то PR-кривая проходит через точку (1,1).

* Отметим также, что кривая характеризует не один конкретный алгоритм $a(x)$, а **множество** алгоритмов, которые отличаются друг от друга выбором значения порога $t$.

В связи с этим в роли метрики качества полученного множества алгоритмов можно рассмотреть **AUC-PR** (Area Under Curve Precision-Recall) — площадь под графиком построенной кривой.

```python
from sklearn.metrics import auc
```

```python
auc(recall, precision)
```

Еще одним способом визуализации зависимости метрик качества от порога, подходящим в этом случае, является *ROC*. По осям: 
 - $FPR = \frac{FP}{FP + TN}$ (*false positive rate*) по оси X
 - $TPR = \frac{TP}{TP + FN}$ (*true positive rate*) по оси Y
 
* ROC проходит через точки (0,0) и (1,1)
* Если при каком-то значении порога $t$ алгоритм $a(x)$ идеально разделяет объекты 2 классов, то ROC проходит через точку (0,1).
 
Аналогично можем рассмотреть в роли метрики качества **AUC-ROC** — площадь под такой кривой. Наихудшая модель имеет площадь 0.5, идеальная — 1.

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:, 1])
plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

```python
auc(fpr, tpr)
```

# Отложенная выборка, кросс-валидация и гиперпараметры

Предположим, что мы решаем задачу восстановления регрессии при помощи линейной модели:

$$a(x) = \langle w, x\rangle$$

В качестве оптимизируемого функционала используется среднеквадратичная ошибка (MSE) с L1-регуляризатором, т.е. решается следующая оптимизационная задача:

$$Q(a, X^\ell) = \frac{1}{\ell} \sum_{i=1}^\ell (a(x_i) - y_i)^2 + \lambda \sum_{j=1}^d \left| w\right| \to \min_w.$$

$\lambda$ является параметром регуляризации. Как его правильно подбирать?

В алгоритмах машинного обучения следует различать параметры модели и гиперпараметры (структурные параметры). Обычно параметры модели настраиваются по обучающей выборке $X^\ell$ в ходе обучения (например, веса $w$), в то время как гиперпараметры задаются заранее (например, $\lambda$).

Давайте рассмотрим варианты, позволяющие подобрать значение $\lambda$ в решаемой задаче. Для начала выберем несколько значений этого параметра, среди которых будем производить выбор, например, {0.01, 0.03, 0.1, 0.3, 1, 3}. После этого для каждого значения $\lambda$ обучим модель на обучающей выборке, вычислим значение оптимизируемого функционала $Q$ для полученного алгоритма и выберем то значение $\lambda$, для которого значение функционала $Q$ будем наилучшим. На каких данных необходимо измерять значение функционала $Q$? Нельзя измерять его на обучающей выборке, поскольку модель уже "подстроилась" под неё. Существует несколько вариантов:

1) Разбить имеющуюся выборку на обучающую и контрольную. Таким образом, используется следующая схема:
* Для каждого значения $\lambda$
    * Обучить модель на обучающей выборке X_train
    * Вычислить $Q$ на контрольной выборке

Однако в этом случае полученное значение $Q$ не будет характеризовать то, как хорошо полученный алгоритм будет работать на новых данных.

2) Для устранения описанной выше проблемы можно разбить данные на 3 непересекающихся подвыборки: обучающую (train), валидационную (validation) и контрольную (test). В этом случае используется следущая схема:
* Для каждого значения $\lambda$
    * Обучить модель на обучающей выборке X_train
    * Вычислить $Q$ на валидационной выборке

После выбора значения $\lambda$ с учетом вычисленных значений $Q$ на валидационной выборке можно использовать тестовую выборку для окончательной оценки качества.

3) Кросс-валидация. В этом случае выборка разделяется на $k$ примерно равных частей (фолдов, folds), каждая из которых поочередно становится контрольной выборкой. В этом случае используется следущая схема:
* Для каждого значения $\lambda$
    * Обучить модель на всех фолдах, кроме первого. Для полученного алгоритма вычисляется $Q$ на объектах первого фолда (обозначим его за $Q_1$).
    * Обучить модель на всех фолдах, кроме второго. Для полученного алгоритма вычисляется $Q$ на объектах второго фолда (обозначим его за $Q_2$).
    * ...
    * Обучить модель на всех фолдах, кроме $K$-ого (последнего). Для полученного алгоритма вычисляется $Q$ на объектах $K$-ого фолда (обозначим его за $Q_K$).
    * Вычислить оценку для текущего $\lambda$ как среднее арифметическое $K$ оценок: $\frac{1}{K} (Q_1 + ... + Q_K)$

![](http://scott.fortmann-roe.com/docs/docs/MeasuringError/crossvalidation.png)

Давайте разберем на примере, как работает кросс-валидация. Загрузим данные и выделим 30% на контрольную выборку.

```python
import pandas as pd
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.preprocessing import scale
from sklearn.metrics import mean_squared_error

data = pd.read_csv('bikes_rent_scaled.csv')
X_train = data.iloc[:, :-1]
y_train = data.iloc[:, -1]
```

```python
X_train.head()
```

```python
y_train
```

Загруженный датасет содержит для каждого дня некоторую календарную информацию, погодные характеристики и пр., по которым необходимо предсказать значение признака cnt — количество прокатов велосипедов в некотором автоматизированном пункте проката.

Будем решать данную задачу при помощи регрессии Lasso. В sklearn уже реализован класс для автоматического подбора параметра регуляризации, поэтому создадим объект-регрессор и обучимся на обучающей выборке.

```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
alphas = np.arange(1, 100, 5) # значения параметра регуляризации, из которых будем выбирать
Reg_LassoCV = LassoCV(alphas = alphas, cv = 5) # cv — количество фолдов при кросс-валидации
Reg_LassoCV.fit(X_train, y_train)
alphas_n = Reg_LassoCV.alphas_  # значения alpha
mse_path = Reg_LassoCV.mse_path_ # значения функционала для каждого фолда для указанных alpha
avg = np.mean(mse_path, axis = 1) # усредненная оценка для каждого alpha

print('\n'.join(['alpha = ' + str(al) + ': MSE = ' + str(mse) for (al, mse) in zip(alphas_n, avg)]))
print('Best alpha:', Reg_LassoCV.alpha_)
plt.plot(alphas_n, avg)
plt.xlabel('alpha')
plt.ylabel('MSE')
```

Как видно, у графика ошибки на кросс-валидации есть характерный минимум, соответствующий оптимальному значению параметра регуляризации.

Давайте посмотрим, что будет, если оценивать ошибку на той же выборке, на которой проводится обучение модели для каждого значения $\alpha$:

```python
alphas = np.linspace(0.1, 10, 100)
mse = []

for alpha in alphas:
    lasso = Lasso(alpha = alpha)
    lasso.fit(X_train, y_train)
    mse.append(mean_squared_error(lasso.predict(X_train), y_train) + alpha * np.linalg.norm(lasso.coef_, ord = 1))
    print('alpha =', alpha, ': MSE = ', mse[-1])

plt.plot(alphas, mse)
plt.xlabel('alpha')
plt.ylabel('MSE')
```

Как мы видим, оптимальное значение параметра регуляризации стремится стать равным нулю. Почему так происходит? (разобрать на доске)
